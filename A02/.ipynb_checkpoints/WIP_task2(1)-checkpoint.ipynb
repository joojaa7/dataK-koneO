```python
from ucimlrepo import fetch_ucirepo 
import pandas as pd
  
# fetch dataset 
breast_cancer_wisconsin_diagnostic = fetch_ucirepo(id=17) 
data = breast_cancer_wisconsin_diagnostic  
# data (as pandas dataframes) 
X = breast_cancer_wisconsin_diagnostic.data.features 

# Testing the model with only the mean values
# Drop all columns that don't have '1' in their name
X_mean = breast_cancer_wisconsin_diagnostic.data.features[
    [col for col in breast_cancer_wisconsin_diagnostic.data.features.columns if '1' in col]
]

y = breast_cancer_wisconsin_diagnostic.data.targets 
  
#X.describe()
```

```python
df = pd.concat([X, y], axis=1)
df
```

Separate data by the diagnosis

```python
df_M = df[df["Diagnosis"]=="M"]
df_B = df[df["Diagnosis"]=="B"]
```

```python
import matplotlib.pyplot as plt

def plot_data(df):
    plt.figure(figsize=(10, 20))
    for i, col in enumerate(df.select_dtypes(include=['int', 'float'])):
        plt.subplot(10, 3, i + 1)
        df[col].plot(kind='hist', title=col)
        plt.xlabel(col)
        plt.tight_layout()
```

```python
plot_data(df_B)
```

```python
plot_data(df_M)
```

Compare the dataset to inspect for significant differences.

```python
desc_df_M = df_M.describe()
desc_df_B = df_B.describe()
desc_df_M.compare(desc_df_B,result_names=('Malignant', 'Benign'),keep_equal=True)
```

Get classifier and split the dataset into a training set and a testing set.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=123)
```

Apply standardization to the numerical values.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X_train), columns=X_train.columns) # preserves column names
X_test_scaled = pd.DataFrame(scaler.transform(X_test), columns=X_test.columns)
X_scaled.describe()
```

Get classifier.


```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=7)
```

Train the classifier on the scaled training data.

```python
knn.fit(X_scaled, y_train.values.ravel())
```

Test the performance and present associated confusion matrix together with the most common evaluation metrics (Accuracy, precision & recall).

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score

y_pred = knn.predict(X_test_scaled.values)

accuracy_test = accuracy_score(y_test, y_pred)
recall_test = recall_score(y_test, y_pred, pos_label='M')
precision_test = precision_score(y_test, y_pred, pos_label='M')

print(f"Accuracy of kNN classifier on the test set: {accuracy_test:.3f}")
print(f"Recall score of kNN classifier on the test set: {recall_test:.3f}")
print(f"Precision score kNN classifier on the test set: {precision_test:.3f}")
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cm = confusion_matrix(y_test, y_pred)
cmd = ConfusionMatrixDisplay(cm, display_labels=['M', 'B'])
cmd.plot()
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn_unscaled = KNeighborsClassifier(n_neighbors=3)
knn_unscaled.fit(X_train, y_train.values.ravel())

y_pred = knn_unscaled.predict(X_test.values)

accuracy_test = accuracy_score(y_test, y_pred)
recall_test = recall_score(y_test, y_pred, pos_label='M')
precision_test = precision_score(y_test, y_pred, pos_label='M')

print(f"Accuracy of kNN classifier on the test set: {accuracy_test:.3f}")
print(f"Recall score of kNN classifier on the test set: {recall_test:.3f}")
print(f"Precision score kNN classifier on the test set: {precision_test:.3f}")
```
